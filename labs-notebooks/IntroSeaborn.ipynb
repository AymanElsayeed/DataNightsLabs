# Introduction to Seaborn
Seaborn is a Python data visualization library based on matplotlib.
It provides a high-level interface for drawing attractive and informative statistical graphics.

* scatter plot
* line plot
* bar plot
* box plot
* violin plot
* heatmap
* pairplot
* jointplot


[Seaborn doc](https://seaborn.pydata.org/index.html)


```python
import seaborn as sns
```

```python
# Load the example flights dataset and convert to long-form
flights_long = sns.load_dataset("flights")

# load iris dataset
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

### Scatter plot

Shows the relationship between two variables

```python

ax = sns.scatterplot(data=iris, x="petal_length", y="petal_width")
ax.set_title("Iris dataset")
ax.set_xlabel("Petal Length")
ax.set_ylabel("Petal Width")
```

```python
ax = sns.scatterplot(data=iris, x="petal_length", y="petal_width", hue="species")
# set legend to the right bottom corner
ax.legend(loc='lower right')
```

### Countplot / BarPlot

shows distribution of categorical variables

Good to compare two categorical variables

```python
ax = sns.countplot(data=iris, x="species")
```

### Histogram

Show the distribution of numeric variable

1. Central tendency
2. Skewness

```python
sns.histplot(data=iris, x="petal_length", bins=90, kde=True, hue="species")
```

### Pair plot

```python
ax = sns.pairplot(data=iris)
```

```python
ax = sns.pairplot(data=iris, hue="species")
```

### HeatMap

```python
ax = sns.heatmap(iris.iloc[:, :-1].corr())
```

### Lineplot

Shows the change overtime

1. Cycical pattern
2. Trend
3. Compare

```python
# Plot the responses for different events and regions in a lineplot
sns.lineplot(data=flights_long, x="year", y="passengers")
```

```python

ax = sns.lineplot(data=flights_long, x="month", y="passengers", hue="year")
```

```python
sns.lineplot(data=iris, x="petal_length", y="petal_width")
```
